<h1 style="font-family: Lucida Console"><b>Missing Value detection and handling:</b>üïµÔ∏è‚Äç‚ôÇÔ∏è</h1>
<ul>
    <li><p style="font-family: times-new-roman;  font-size: 17px;"> Missing value  for <span style="color: #2e4053; font-weight:bold">imd_band</span> is found in <i>studentInfo</i> dataset. The missing values are replaced by the <i>most frequent</i> imd_band value in every region.
        <li><p style="font-family: times-new-roman;  font-size: 17px;">For <i>week_from and week_to columns</i> in Vle dataset, around <b>82 %</b> of the values is missing, therefore we  <span style="color:  #f75144 ; font-weight:bold">drop</span> both the columns from the dataset

<div class="alert success-alert" style="font-family: times-new-roman">
  <h4 style="font-family: times-new-roman"><center>üëÄ There is no missing value found in the studentVle dataset
</center></h4>
</div>

```python
import missingno as mn
mn.matrix(studentInfo, figsize=(10,6))
```

```python
mn.matrix(studentVle, figsize=(10,6))
```

```python
mn.matrix(vle, figsize=(10,6))
```

```python
studentInfo['imd_band_2'] = studentInfo.apply(lambda x: studentInfo[studentInfo['region']==x['region']]['imd_band'].mode()[0] \
    if pd.isna(x['imd_band']) else x['imd_band'], axis=1)
```

```python
# Dropping the imd_band variable after imputing the missing value with the most frequent value per region
new_studentInfo= studentInfo.drop(['imd_band'],axis=1)
```

```python
new_vle=vle.drop(['week_from','week_to'],axis=1)
```

```python
Vle_activity=pd.merge(studentVle,new_vle,on=['code_module','code_presentation','id_site'],how='outer')
```

```python
merged_Df=pd.merge(Vle_activity,new_studentInfo,on=['code_module','code_presentation','id_student'],how='outer')
```

```python
#Q1
# total clicks per activity type 
total_click_per_activity=Vle_activity\
.groupby(['activity_type'])\
.agg(total_click=('sum_click',sum)).reset_index()
```

```python
#Q1
fig=px.bar(total_click_per_activity, x="activity_type", y="total_click",color='activity_type',title='Total Clicks per activity(Type of content)',text_auto=True)
fig.show()
```
